```python
bpa_tim_varios_anos = pd.DataFrame()

for ano in range(2021,2024):
  # # Ler informações de arquivo (.csv) específico
  # Corrected file path using os.path.join
  file_path = os.path.join(pasta_drive, 'dfp', f'dfp_cia_aberta_BPA_con_{ano}.csv')
  bpa = pd.read_csv(file_path, sep=';', encoding='ISO-8859-9', decimal=',')

  # # Filtrar informações da coluna ORDEM_EXERC == ÚLTIMO
  bpa = bpa[bpa['ORDEM_EXERC'] == 'ÚLTIMO']

  # # Filtrar empresa por CNPJ
  bpa_tim = bpa[ bpa['CNPJ_CIA'] == '02.421.421/0001-11']

  # # Filtrar apenas as colunas que serão utilizadas
  bpa_tim = bpa_tim[['DT_REFER','CD_CONTA','DS_CONTA','VL_CONTA']]

  # # Converter valores numéricos
  bpa_tim['VL_CONTA'] = pd.to_numeric(bpa_tim['VL_CONTA'])

  # # Reduzir as contas
  bpa_tim = bpa_tim[bpa_tim['CD_CONTA'].str.len() <= 7]

  # # Reiniciar os indices
  bpa_tim.reset_index(inplace=True, drop=True)

  # # Obter o Ativo Total
  filtered_bpa_tim = bpa_tim[bpa_tim['CD_CONTA'] == '1']
  if not filtered_bpa_tim.empty:
    ativo = filtered_bpa_tim['VL_CONTA'].iloc[0]
  else:
    # Handle the case where 'CD_CONTA' with value '1' is not found
    # You can print a warning, assign a default value, or skip the iteration
    print(f"Warning: 'CD_CONTA' with value '1' not found for year {ano}")
    continue  # Skip to the next year

# # Criar coluna (%)
  # bpa_tim['%'] = bpa_tim['VL_CONTA'] / ativo * 100

  # transformação dos indices, colunas e linhas
  bpa_tim = bpa_tim.pivot(index=['CD_CONTA','DS_CONTA'], columns='DT_REFER', values='VL_CONTA')

  # criar (%) com iloc (valores de indices das colunas)
  bpa_tim['%'] = bpa_tim.iloc[:,0:] / ativo * 100

  # Agrupamento por colunas
  bpa_tim_varios_anos = pd.concat([bpa_tim_varios_anos, bpa_tim], axis=1)
  # display(bpa_tim.head())
  # break
bpa_tim_varios_anos
```

```python
# Criar DataFrame com todas as informações do anos coletados
demonstracoes_anuais = pd.DataFrame()

pasta_drive2 = '/content/drive/MyDrive/banco_dados/cvm/dfp/'

# Realiza repetições dos códigos
for ano in range(2015,2024):
  # Leitura dos arquivos (.csv)
  bpa = pd.read_csv(pasta_drive2+f'/dfp_cia_aberta_BPA_ind_{ano}.csv', sep=';', encoding='ISO-8859-9', decimal=',')

  # Filtrar apenas informações do ORDEM_EXERC == ÚLTIMO
  bpa = bpa[bpa['ORDEM_EXERC'] == 'ÚLTIMO']

  # Converter dados numéricos e data
  bpa['VL_CONTA'] = pd.to_numeric(bpa['VL_CONTA'])
  bpa['DT_REFER'] = pd.to_datetime(bpa['DT_REFER'])

  # Criar coluna ANO
  bpa['ANO'] = bpa['DT_REFER'].dt.year


  # Leitura dos arquivos (.csv)
  dre = pd.read_csv(pasta_drive2+f'dfp_cia_aberta_DRE_ind_{ano}.csv', sep=';', encoding='ISO-8859-9', decimal=',')

  # Filtrar apenas informações do ORDEM_EXERC == ÚLTIMO
  dre = dre[dre['ORDEM_EXERC'] == 'ÚLTIMO']

  # Converter dados numéricos e data
  dre['VL_CONTA'] = pd.to_numeric(dre['VL_CONTA'])
  dre['DT_REFER'] = pd.to_datetime(dre['DT_REFER'])

  # Criar coluna ANO
  dre['ANO'] = dre['DT_REFER'].dt.year

  # Agrupar por linha
  demonstracoes_anuais = pd.concat([demonstracoes_anuais,bpa])
  demonstracoes_anuais = pd.concat([demonstracoes_anuais,dre])
```
